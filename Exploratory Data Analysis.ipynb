```python
##Preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder
label= LabelEncoder()
```

```python
df1=df.copy()
```

```python
df1['workclass']=label.fit_transform(df1['workclass'])
df1['education']=label.fit_transform(df1['education'])
df1['marital_status']=label.fit_transform(df1['marital_status'])
df1['occupation']=label.fit_transform(df1['occupation'])
df1['relationship']=label.fit_transform(df1['relationship'])
df1['race']=label.fit_transform(df1['race'])
df1['sex']=label.fit_transform(df1['sex'])
df1['native_country']=label.fit_transform(df1['native_country'])
df1

```

```python
##Isolation Forest
```

```python
from sklearn.ensemble import IsolationForest
```

```python
features = ['workclass','education','marital_status','occupation','relationship','race','sex','native_country']
```

```python
iso_forest=IsolationForest(contamination=0.01,random_state=42)
```

```python
outlaiers=iso_forest.fit_predict(df1[features])
```

```python
outlaiers
```

```python
df1['outlaiers'] = outlaiers
```

```python
clean_data = df1[df1['outlaiers'] != -1]
clean_data
```

```python
df1.outlaiers.value_counts()
```

```python
X = clean_data.drop(columns=['income', 'outlaiers'])
X
```

```python
y = clean_data['income']
y
```

```python
label_encoder = LabelEncoder()
label_encoder
y_encoded = label_encoder.fit_transform(y)
y_encoded
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=1000, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

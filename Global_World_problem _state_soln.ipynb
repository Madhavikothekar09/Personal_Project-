```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ','',regex=True)
df.columns = df.columns.str.replace('[()]','_',regex=True)
df.columns = df.columns.str.replace('LandArea_Km2_','LandArea_Km2',regex=True)
df.columns = df.columns.str.replace('Density\n_P/Km2_','Density',regex=True)
df.columns = df.columns.str.replace('AgriculturalLand_%_','AgriculturalLand',regex=True)
df.columns = df.columns.str.replace('Population:Laborforceparticipation_%_','Population:Laborforceparticipation',regex=True)
df.columns = df.columns.str.replace('Taxrevenue_%_','Taxrevenue',regex=True)
df.columns = df.columns.str.replace('Co2-Emissions','Co2_Emissions',regex=True)
df.columns = df.columns.str.replace('Population: Labor force participation (%)','Labor_force_ participation',regex=True)

                                    

```

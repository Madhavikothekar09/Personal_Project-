```python
print(df.type_of_private_residential.unique())
a = input('Enter Type of Private Residential:')
b = df[df.type_of_private_residential == a].reset_index()
b
```

```python
sb.lineplot(x=b.financial_year,y=b.no_of_cases)
plt.title('Number of Cases over the years')
plt.xlabel('Financial Year')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
sb.barplot(x=b.financial_year, y=b.median_annual_value,color='#243848',edgecolor='grey')
plt.title('Median annual value over the years')
plt.xlabel('Financial Year')
plt.ylabel('Median Annual Value')
plt.xticks(rotation=90)
plt.show()
```

```python
sb.barplot(x=b.financial_year, y=b.property_tax_collection,color='#243848',edgecolor='grey')
plt.title('Property Tax Collection over the years')
plt.xlabel('Financial Year')
plt.ylabel('Property Tax Collection')
plt.tick_params(axis='x',rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sb.boxplot(x='type_of_private_residential', y='median_annual_value' ,data=b)
plt.title('Distribution of Median Annual Value by Private Residential')
plt.xlabel('Private Residential')
plt.ylabel('Median Annual Value')

plt.subplot(1,2,2)
sb.boxplot(x='type_of_private_residential', y='property_tax_collection',data=b)
plt.title('Distribution of Property Tax Collection by Private Residential')
plt.xlabel('Private Residential')
plt.ylabel('Property Tax Collection')
plt.show()
```

```python
corr_matrix = df[['no_of_cases','median_annual_value','property_tax_collection']].corr()
plt.figure(figsize=(10,8))
sb.heatmap(corr_matrix,annot=True,cmap='Blues',fmt='.2f')
plt.title('Correlation Heatmap')   
plt.show()           
```

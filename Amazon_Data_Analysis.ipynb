```python
df.columns = df.columns.str.replace('State-Province','State_Province',regex=True)
df.columns = df.columns.str.replace('BirthDate','Birth_Date',regex=True)
df.columns = df.columns.str.replace('EmailAddress','Email_Address',regex=True)
df.columns = df.columns.str.replace('YearlyIncome','Yearly_Income',regex=True)
df.columns = df.columns.str.replace('CategoryName','Category_Name',regex=True)
df.columns = df.columns.str.replace('ItemCost','Item_Cost',regex=True)
df.columns = df.columns.str.replace('NumberofItem','Number_of_Item',regex=True)
df.columns = df.columns.str.replace('MaritalStatus','Marital_Status',regex=True)
```
